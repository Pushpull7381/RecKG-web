```python
import pandas as pd


df = pd.read_csv('sample_ym_kg_neo4j_data.txt', delimiter=',') 
# df = df.apply(lambda col: col.str.split("|") if col.dtype == "object" else col)

print(df.head())
```

```python
df.drop(columns=["distributor", "genres", "directors", "crew", "actors", "age", "release_year", "crew_types"])
```

```python
import json

# Create nodes for users and movies
user_nodes = [{'id': user_id, 'type': 'user', 'data': {'label': user_id}} for user_id in df['userId'].unique()]
movie_nodes = [{'id': str(movie_id), 'type': 'movie', 'data': {'label': title}} for movie_id, title in df[['movieId', 'title']].drop_duplicates().values]

# Create edges for ratings
edges = [{'id': f"{row['userId']}-{row['movieId']}", 'source': row['userId'], 'target': str(row['movieId']), 'data': {'rating': row['rating']}} for _, row in df.iterrows()]

# Combine nodes and edges
graph = {
	'nodes': user_nodes + movie_nodes,
	'edges': edges
}

# Convert to JSON
graph_json = json.dumps(graph, indent=2)

# Save to a file
with open('graph.json', 'w') as f:
	f.write(graph_json)

print(graph_json)
```
